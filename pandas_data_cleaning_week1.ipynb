```python
import pandas as pd

# Step 1: Load dataset
file_path = "jeans.csv"  

df = pd.read_csv(file_path)
print("Dataset loaded successfully")

# Step 2: Explore dataset
print("\nFirst 5 rows")
print(df.head())

print("\nLast 5 rows")
print(df.tail())

print("\nShape:", df.shape)

print("\nColumns:", df.columns)

print("\nData types:")
print(df.dtypes)

# Step 3: Clean final_price properly
df['final_price'] = (
    df['final_price']
    .astype(str)
    .str.replace(r'[‚āĻ,"]', '', regex=True)
    .str.strip()
)

df['final_price'] = pd.to_numeric(df['final_price'], errors='coerce')

# Step 4: Handle missing values
df['rating'] = pd.to_numeric(df['rating'], errors='coerce')
df['rating'] = df['rating'].fillna(df['rating'].median())

df['discount'] = pd.to_numeric(df['discount'], errors='coerce')
df['discount'] = df['discount'].fillna(0)

print("\nMissing values handled")

# Step 5: Basic analysis
print("\nSelected columns")
print(df[['final_price', 'rating']].head())

print("\nProducts with rating > 4")
print(df[df['rating'] > 4].head())

# Step 6: Remove duplicates
before = df.shape
df = df.drop_duplicates()
after = df.shape

print("\nBefore duplicates:", before)
print("After duplicates:", after)

# Step 7: Derived column
df['quantity'] = 2

# safe calculation
df['final_price'] = df['final_price'].fillna(0)
df['total_amount'] = df['final_price'] * df['quantity']

print("\nTotal amount sample")
print(df[['final_price', 'quantity', 'total_amount']].head())

# Step 8: Insights
print("\nBasic insights")

print("Average rating:", df['rating'].mean())
print("Max price:", df['final_price'].max())
print("Min price:", df['final_price'].min())

# Step 9: Save cleaned file (GitHub-safe)
df.to_csv("cleaned_shopping_data.csv", index=False)

print("\nCleaned dataset saved successfully")
```
